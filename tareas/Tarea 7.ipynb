## Ejercicio 1

A partir del siguiente código, intenta obtener el gráfico del dibujo llamado ejercicio 1 en el enunciado.

Pista: investiga el parámetro fill_between de matplotlib.

n = 256

X = np.linspace(-np.pi, np.pi, n, endpoint=True)

Y = np.sin(2 * X) 

plt.plot(X, Y + 1, color='blue', alpha=1.00)

plt.plot(X, Y - 1, color='blue', alpha=1.00)

```python
n = 256
X = np.linspace(-np.pi, np.pi, n, endpoint=True)
Y = np.sin(2 * X) 

plt.plot(X, Y + 1, color='blue', alpha=1.00)
plt.fill_between(X,Y+1,1,alpha=0.6)

plt.plot(X, Y - 1, color='blue', alpha=1.00)
plt.fill_between(X,Y-1,-1,where=[(Y-1>=-1) for Y in Y], alpha=0.6)
plt.fill_between(X,Y-1,-1,where=[(Y-1<=-1) for Y in Y], alpha=0.6)
```

## Ejercicio 2

A partir del siguiente código, intenta obtener el gráfico del dibujo llamado ejercicio 2 en el enunciado.

Pista: el color viene dado por el ángulo que forman x e y con la horizontal

n = 1024

Y = np.random.normal(0,1,n) 

plt.scatter(X,Y)

```python
n = 1024
X = np.random.normal(0, 1, n)
Y = np.random.normal(0, 1, n)

T = np.arctan2(Y, X)

plt.axes([0.025, 0.025, 0.95, 0.95])
plt.scatter(X, Y, s=75, c=T, alpha=.5)
plt.xlim(-1.5, 1.5)
plt.xticks(())
plt.ylim(-1.5, 1.5)
plt.yticks(())

plt.show()
```

n = 12

Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)

```python
n = 12
X = np.arange(n)
Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)

plt.axes([0.025, 0.025, 0.95, 0.95])
plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')
plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white')

for x, y in zip(X, Y1):

    plt.text(x + 0.4, y + 0.05, '%.2f' % y, ha='center', va= 'bottom')

for x, y in zip(X, Y2):

    plt.text(x + 0.4, -y - 0.05, '%.2f' % y, ha='center', va= 'top')

plt.xlim(-.5, n)
plt.xticks(())
plt.ylim(-1.25, 1.25)
plt.yticks(())

plt.show()
```

## Ejercicio 4

A partir del siguiente código, intenta obtener el gráfico del dibujo llamado ejercicio 4 en el enunciado.

Pista: tendrás que modificar la Z.

Z = np.random.uniform(0, 1, 20)

plt.pie(Z)

```python
n = 20
Z = np.ones(n)
Z[-1] *= 2

plt.axes([0.025, 0.025, 0.95, 0.95])
plt.pie(Z, explode=Z*.05, colors = ['%f' % (i/float(n)) for i in range(n)])
plt.axis('equal')
plt.xticks(())
plt.yticks()

plt.show()
```
